# Time Series

### Creating Datetime

```python
dates = pd.date_range(start='2024-01-01', periods=10, freq='D')
```

```python
import numpy as np
```

```python
ts = pd.DataFrame({
    'Date': dates,
    'Sales': np.random.randint(100,500,10)
})
```

```python
ts
```

```python
ts['Date']=pd.to_datetime(ts['Date'])
```

```python
ts.set_index('Date',inplace=True)
```

```python
ts
```

```python
ts['Year']=ts.index.year
```

```python
ts
```

```python
ts['Month']=ts.index.month
ts['Weekday']=ts.index.day_name()
ts
```

```python
ts.resample('D').sum()
```

```python
dates= pd.date_range(start='2026-06-01', periods=10, freq='D')
ts2=pd.DataFrame({
    'Date':dates,
    'Sales': np.random.randint(100,500,10)
})
ts2['Date']=pd.to_datetime(ts2['Date'])
ts2.set_index('Date', inplace=True)
ts2
```

```python
ts2.resample('W').sum()
```

```python
ts2['7Day_Avg']=ts2['Sales'].rolling(window=7).mean()
```

```python
ts2
```

```python
ts2['Previous_day']= ts2['Sales'].shift(1)
ts2
```

## Reshaping

### melt

```python
wide = pd.DataFrame({
    'Name': ['Rahim', 'Karim'],
    'Math': [80, 90],
    'Science': [85, 88]
})
```

```python
wide
```
