# Sales Analysis Dashboard Data Prep

```python
import pandas as pd
```

```python
sales = pd.DataFrame({
    'Date': pd.date_range('2024-01-01', periods=12, freq='ME'),
    'Region': ['Dhaka','Ctg','Sylhet']*4,
    'Sales': [50000, 30000, 20000, 55000, 32000, 22000,
              58000, 35000, 25000, 60000, 38000, 27000],
    'Cost': [30000, 20000, 15000, 32000, 21000, 16000,
             33000, 22000, 17000, 34000, 23000, 18000]
})

sales['Profit'] = sales['Sales'] - sales['Cost']
sales['Profit_Margin'] = (sales['Profit'] / sales['Sales'] * 100).round(2)

report = sales.groupby('Region').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Avg_Margin=('Profit_Margin', 'mean')
).sort_values('Total_Profit', ascending=False)

report
```
