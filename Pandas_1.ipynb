# Data Transformation

```python
df = pd.DataFrame({
    'Name': ['Rahim', 'Karim', 'Salma'],
    'Salary': [40000, 45000, 30000],
    'Bonus_Percent': [10, 15, 5]
})
```

```python
df['Bonus'] = df['Salary']*df['Bonus_Percent']/100
```

```python
df
```

```python
df['Total'] = df['Salary']+df['Bonus']
df
```

```python
def tax_bracket(salary):
    if salary > 40000:
        return 'High'
    else:
        return 'Low'

df['Tax_Bracket'] = df['Salary'].apply(tax_bracket)   

df
```

```python
df.sort_values('Salary', ascending=False)
```

```python
df.sort_values(['Bonus','Salary'], ascending=[True,False])
```

```python
df.rename(columns={'Salary':'Monthly_Salary'})
```

## GroupBy & Aggregation

```python
sales = pd.DataFrame({
    'Region': ['Dhaka', 'Ctg', 'Dhaka', 'Sylhet', 'Ctg', 'Dhaka'],
    'Product': ['A', 'B', 'A', 'A', 'B', 'B'],
    'Sales': [1000, 1500, 1200, 800, 2000, 900],
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Jan', 'Feb']
})
```

```python
sales
```

```python
sales.groupby('Region')['Sales'].sum()
```

```python
sales.groupby('Region')['Sales'].agg(['sum','mean','count','max'])
```

```python
sales.groupby(['Region','Product'])['Sales'].sum()
```

```python
sales.groupby('Region').agg(
    Total_Sales=('Sales','sum'),
    Avg_Sales=('Sales','mean')
)
```

```python
pd.pivot_table(sales, values='Sales', index='Region', columns= 'Month', aggfunc='sum', fill_value=0)
```

## Merging

```python
employees = pd.DataFrame({
    'EmpID': [1, 2, 3, 4],
    'Name': ['Rahim', 'Karim', 'Salma', 'Nusrat'],
    'DeptID': [10, 20, 10, 30]
})
employees
```

```python
departments = pd.DataFrame({
    'DeptID': [10, 20, 30],
    'DeptName': ['Sales', 'IT', 'HR']
})
departments
```

```python
merged = pd.merge(employees, departments, on='DeptID', how='left')
```

```python
merged
```
